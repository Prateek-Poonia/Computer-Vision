```python
import mahotas
import numpy as np
from pylab import gray, imshow, show
import os

#loading image
img = mahotas.imread("/content/images.jpg")

#filtering image
img = img[:,:,0]

#otsu method
t_otsu = mahotas.otsu(img)

#img values should be greater than otsu value
img =  img > t_otsu

print("image threshold using otsumethod ")
imshow(img)
show()
```

```python
#crop to bbox
new_img = mahotas.croptobbox(img)
```

```python
imshow(new_img)
show()
```

```python
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2gray
import numpy as np
```

```python
#load the uploadded image
image_path = "/content/images.jpg"

#convert to grayscale
grayscale = rgb2gray(imread(image_path))

#scale grayscale values to range(0,255)
grayscale = (grayscale*255).astype(np.uint8)
```

```python
#apply log transformation
c = 255/np.log(1+np.max(grayscale))
transformed_image = c*np.log(1+grayscale)
transformed_image = transformed_image.astype(np.uint8)
```

```python
#sample grayscale image (2D Numpy array)
grayscale = np.array([[100,150],[200,250]])

def transform_function(r):
  s = 2*r # transformation : multiply each pixel by 2
  return s
```

```python
#apply the transformation
transformed_image = transform_function(grayscale)
```

```python
#display the ttransformed image
plt.imshow(transformed_image, cmap='gray')
plt.title('Transformed image')
plt.show()
```
