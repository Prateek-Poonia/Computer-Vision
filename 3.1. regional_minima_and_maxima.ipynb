```python
import mahotas
import numpy as np
import os
import matplotlib.pyplot as plt
from pylab import gray, imshow, show


#load the image/content/images.jpg
img = mahotas.imread("/content/images.jpg")

```

```python
#filtering the image
img = img[:,:,0]
```

```python
print("image")

#showing the image
imshow(img)
show()

#apply the regional minima
new_img = mahotas.regmin(img)

#showing the minima image
print("regional minima imge")
imshow(new_img)
show()
```

```python
#apply the regional maxima
new_img = mahotas.regmax(img)

#showing the maxima image
print("regional maxima imge")
imshow(new_img)
show()
```

```python
#finding the mean value
mean = np.mean(img)
mean=img.mean()
```
